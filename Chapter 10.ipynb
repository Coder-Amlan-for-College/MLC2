```python
import numpy as np
```

```python
#inputs
x=np.array([0.5,0.2,0.1])

#weights
w = np.array([0.4,-0.7,0.2])

#weighted sum
y_sum = np.dot(x,w)

#simple activation (identity)
output=y_sum

print("Weighted Sum:",y_sum)
print("Output:",output)
```

```python
def biological_neuron(signals,threshold=1.0):
    soma=sum(signals)
    if soma>=threshold:
        return 1
    return 0

signals=[0.3,0.4,0.5]
print("Neuron Output:",biological_neuron(signals))
```

```python
import numpy as np
def artificial_neuron(inputs,weights,bias=0):
    y_sum=np.dot(inputs,weights)+bias
    return y_sum

x=np.array([1,0,1])
w=np.array([0.5,-0.6,0.2])

print("Summation Output:",artificial_neuron(x,w,bias=0.1))
```

```python
import math

#Identity Function
def identity(x):
    return x

#Step/Threshold 
def step(x):
    return 1 if x>=0 else 0

#ReLU function
def threshold(x,theta=0.5):
    return 1 if x==theta else 0

#ReLU function
def relu(x):
    return max(0,x)

#Sigmoid Function
def sigmoid(x,k=1):
    return 1/(1+math.exp(-k*x))

#Bipolar Sigmoid
def bipolar_sigmoid(x):
    return (2/(1+math.exp(-x)))-1

#Hyperbolic Tangent
def tanh_n(x):
    return math.tanh(x)

#---------Test Values--------
values=[-2,-1,0,0.5,1,2]

print("Input Values:",values)
print("\n:----Identity Function----")
for v in values:
    print(v,"->",identity(v))

print("\n:---Step Function----")
for v in values:
    print(v,"->",step(v))

print("\n:----Threshold Function (threshold=0.5)----")
for v in values:
    print(v,"->",threshold(v))


print("\n:----ReLU Function----")
for v in values:
    print(v,"->",relu(v))

print("\n:----Sigmoid Function----")
for v in values:
    print(v,"->",round(sigmoid(v),4))

print("\n:----Bipolar Sigmoid Function----")
for v in values:
    print(v,"->",round(bipolar_sigmoid(v),4))

print("\n:----tanh Function----")
for v in values:
    print(v,"->",round(tanh_n(v),4))
```
